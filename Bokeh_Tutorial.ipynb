```
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
```

```
#Prepare some sample data
x = [x for x in range(1,8)]
y = [6, 7, 2, 4, 5, 12, 8]

#Output into this cell
output_notebook()

#Create a new plot with a title and axis labels

p = figure(title="Awesome Bokeh Chart", 
    x_axis_label='day', y_axis_label='temp')

#Adding a line renderer with legend and line thickness
p.line(x, y, legend = "Temp.", line_width=2)

#show the results
print(x)
show(p)
   
```

Reproduction of Hans Rosling analysis of Life Expectancy within given coutnries
